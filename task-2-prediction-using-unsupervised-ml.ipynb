# STEP-1 IMPORTING LIBRARIES

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head() # See the first 5 rows
```

# STEP-8 FINDING THE OPTIMUM NUMBER OF               K-MEANS CLASSIFICATION

```python
x = iris_df.iloc[:,:].values
from sklearn.cluster import KMeans
square_error = []
for K in range(1,11):
    kmeans = KMeans(n_clusters = K)
    kmeans.fit(x)
    square_error.append(kmeans.inertia_)
```

# STEP-9 PLOTTING THE RESULTS ONTO A LINE GRAPH, ALLOWING US TO OBSERVE 'THE ELBOW'

```python
sns.set(style='darkgrid')
plt.figure(figsize=(12,6))
plt.plot(range(1,11), square_error,marker='o',c='blue',mfc='red',ms='10',linewidth='5')
plt.title('The Elbow method',size=20)
plt.xlabel('Number of clusters (K)',size=15)
plt.ylabel('Sum of Square Error ', size=15)
plt.show()
```

**you can clearly see why it is called "The Elboe Method" from the above graph, the optimum clusters is where the elbow occurs. from this we choose the number of cluster as 3**

# STEP-10 APPLYING K-MEANS TO THE DATASET AND CREATING THE K-MEANS CLASSIFIER

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(x)
y_predicted
```

# STEP-11 VISUALISING THE CLUSTERS

```python
sns.set(style='dark')
plt.figure(figsize=(15,6))
plt.scatter(x[y_predicted == 0,0],x[y_predicted == 0,1], s = 70, c = 'yellow', label = 'Iris-setosa')
plt.scatter(x[y_predicted == 1,0],x[y_predicted == 1,1], s = 70, c = 'green', label = 'Iris-versicolour')
plt.scatter(x[y_predicted == 2,0],x[y_predicted == 2,1], s = 70, c = 'blue', label = 'Iris-viginica')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s = 150, c = "red", label = 'centroids')
plt.legend()

```

# now we are going to update the dataframe

```python
iris_df['cluster'] = y_predicted
iris_df['cluster'].value_counts()
```
